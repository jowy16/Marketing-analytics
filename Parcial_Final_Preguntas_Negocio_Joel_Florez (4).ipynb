## 3. Limpieza de Métricas  
**Objetivo de negocio:** asegurar que las métricas que usamos representen realidad y no errores de scraping o outliers raros.

### 3.1. Preguntas de negocio

1. **Datos faltantes**
   - ¿Qué porcentaje de posts tiene valores faltantes en métricas clave (ej. `likes`, `comments`, `views`)?  
   - Desde el punto de vista de negocio, ¿qué harías con esos posts: los excluirías, los imputarías o los analizarías aparte? ¿Por qué?

2. **Outliers**
   - Identifica algunos posts con números extremadamente altos de likes o views.  
   - ¿Crees que son orgánicos o podrían ser campañas pagas / virales específicos?  
   - ¿Cómo cambia tu lectura si incluyes vs. excluyes estos outliers?

3. **Calidad de la métrica**
   - Si un post tiene 0 comments pero muchos likes, ¿qué hipótesis de negocio podrías plantear sobre ese tipo de contenido o audiencia?  
   - ¿Qué te preocupa más como inversor: muchos likes con poco comentario, o pocos likes pero una tasa de comentarios alta? Explica.


```python
columnas_clave = ["commentsCount", "viewsCount"]
faltantes = df[columnas_clave].isna().any(axis=1).sum()
total_posts = df.shape[0]
porcentaje_faltantes = (faltantes / total_posts) * 100

print(f"Porcentaje de posts con valores faltantes en métricas clave: {porcentaje_faltantes:.2f}%")

print("-----------------------------")

umbral_comments = df['commentsCount'].quantile(0.95)
umbral_views = df['viewsCount'].quantile(0.95)

print(f"Valor umbral para commentsCount: {umbral_comments}")
print(f"Valor umbral para viewsCount: {umbral_views}")
outliers = df[(df['commentsCount'] > umbral_comments) | (df['viewsCount'] > umbral_views)]

outliers[['brand', 'commentsCount', 'viewsCount', 'publishedAt']]

```

### **Respuestas**

1. Datos faltantes

Desde una perspectiva de negocio, cuando un post presenta métricas incompletas no es ideal eliminarlo sin más, ya que podríamos perder información relevante para entender el comportamiento de la marca. Sin embargo, imputar valores también puede ser problemático porque introduce supuestos que pueden alterar la interpretación real del desempeño.
La mejor práctica es analizar estos casos por separado: identificar por qué faltan los datos, determinar si se repite en ciertos formatos o tipos de publicaciones y, con base en ello, decidir cómo incorporarlos al análisis. Lo importante es mantener transparencia sobre las limitaciones del dataset y evitar conclusiones basadas en información distorsionada.

2. Outliers

Sobre los posts con valores excepcionalmente altos, es probable que no sean completamente orgánicos. Pueden estar influenciados por campañas pagas, promociones específicas o momentos en los que un contenido se vuelve viral. Detectar esto es esencial porque pueden alterar la percepción del rendimiento habitual de la marca.

Respecto a incluir o excluir estos outliers:

Si los incluyes, aumentan artificialmente los promedios de interacción, dando la impresión de que la marca tiene un desempeño más alto de lo que realmente ocurre en la mayoría de los casos.

Si los excluyes, el análisis se vuelve más representativo del comportamiento típico, aunque corres el riesgo de pasar por alto picos de éxito importantes que también forman parte de la estrategia de la marca.

Lo más equilibrado es presentar ambos escenarios: un análisis general y otro filtrando outliers, para ofrecer una visión completa del rendimiento.

3. Calidad de la métrica

Un post con muchos likes pero sin comentarios suele indicar que el contenido es fácil de consumir o visualmente atractivo, pero no motiva conversación ni participación activa. Puede tratarse de contenido más superficial, informativo o promocional que invita a reaccionar, pero no a interactuar.

Para un inversor, esto puede ser una señal de alerta: muchos likes no siempre implican un engagement profundo. Muestran aprobación, pero no necesariamente conexión o comunidad.
En contraste, un post con pocos likes pero varios comentarios podría indicar un público más involucrado, dispuesto a interactuar y aportar, lo cual puede ser un indicador de una comunidad más sólida y valiosa para campañas a largo plazo.

## 4. Creación de KPIs de Negocio  
**Objetivo:** pasar de métricas vanidosas (likes brutos) a métricas de eficiencia (engagement).

Supongamos que creas un indicador como:

\>
\> `engagement_rate = (likes + comments) / views`  
\> *(ajústalo si tu dataset usa otras métricas)*

### 4.1. Preguntas de negocio

1. **Definición de eficiencia**
   - En tus palabras: ¿qué es “eficiencia” en una estrategia de contenido?  
   - ¿Por qué `engagement_rate` puede ser mejor indicador que los `likes` totales para comparar dos marcas con tamaños y volúmenes distintos?

2. **Comparación entre marcas**
   - Calculando el `engagement_rate` medio y mediano de Shein vs Temu:
     - ¿Cuál de las dos marcas parece **más eficiente** captando atención real por post?  
     - Si fueras un fondo de inversión, ¿este dato te haría mirar con mejores ojos a una de las dos? ¿Por qué?

3. **Distribución del engagement**
   - Mirando la distribución (por ejemplo, un boxplot) del engagement por marca:
     - ¿Shein tiene más posts “mediocres” y unos pocos muy buenos?  
     - ¿Temu tiene menos posts pero más consistentes?  
     - ¿Qué tipo de distribución preferirías si fueras CMO?

4. **Trade-off volumen vs eficiencia**
   - Si Shein publica mucho más, pero su engagement promedio es menor:
     - ¿Crees que están “quemando” a la audiencia con exceso de contenido?  
     - ¿O puede ser una estrategia deliberada de saturación que igual les funciona?


```python
if 'likesCount' not in df.columns:
    df['likesCount'] = 0
df['engagement_rate'] = (df['likesCount'] + df['commentsCount']) / df['viewsCount']
df[['brand', 'likesCount', 'commentsCount', 'viewsCount', 'engagement_rate']].head()



engagement_promedio = df.groupby('brand')['engagement_rate'].mean()
print("Engagement promedio por marca:\n", engagement_promedio)

top_engagement = df.sort_values('engagement_rate', ascending=False).head(5)
print("\nPosts con mayor engagement:\n", top_engagement[['brand', 'likesCount', 'commentsCount', 'viewsCount', 'engagement_rate']])

low_engagement = df.sort_values('engagement_rate', ascending=True).head(5)
print("\nPosts con menor engagement:\n", low_engagement[['brand', 'likesCount', 'commentsCount', 'viewsCount', 'engagement_rate']])


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.boxplot(x='brand', y='engagement_rate', data=df)
plt.title('Distribución del Engagement Rate por Marca')
plt.ylabel('Engagement Rate')
plt.show()


posts_por_marca = df['brand'].value_counts()
print("Número de posts por marca:\n", posts_por_marca)
```

### **Respuestas**

1. Definición de eficiencia

En una estrategia de contenido, la eficiencia mide qué tan bien cada publicación convierte su alcance en interacción real. No basta con publicar en grandes cantidades o acumular muchos likes totales: lo importante es qué porcentaje de las personas que ven un contenido realmente reaccionan a él, ya sea con likes, comentarios o compartidos.
Por eso, el engagement_rate es una métrica más confiable que los likes absolutos al comparar marcas con volúmenes y tamaños distintos. Al relacionar las interacciones con las vistas, permite evaluar la efectividad real de cada publicación y detectar qué marca logra conectar mejor con su audiencia sin importar cuántos posts publica.

2. Comparación entre marcas

Según los datos analizados, Temu resulta ser la marca más eficiente generando interacción por post. Presenta un engagement rate promedio de 0.001017, y sus publicaciones con mayor rendimiento muestran una relación sólida entre vistas e interacciones.
En contraste, Shein aparece con valores NaN, lo que probablemente indica que la métrica de likes (likesCount) no se incorporó correctamente en el cálculo, impidiendo medir su eficiencia.
Si yo fuera un inversionista, este resultado me llevaría a valorar más el desempeño de Temu: aunque publica menos, cada post genera interacción concreta y proporcional a su alcance, lo que refleja una conexión más genuina con la audiencia. Es un indicador de calidad de engagement más fiable que los totales acumulados.

3. Distribución del engagement

Shein no aparece representada en la gráfica debido a la falta de datos necesarios para calcular su engagement, así que no es posible analizar su distribución.

Temu, por su parte, muestra una distribución estable: la caja está compacta y la mayor parte de las publicaciones se concentra en un nivel medio de engagement, con pocos valores extremos.

Para un CMO, esta consistencia es valiosa. Implica que incluso sin muchos posts, Temu mantiene un rendimiento mínimo asegurado en cada publicación, reduciendo el riesgo y asegurando resultados constantes en su estrategia de contenido.

4. Trade-off: volumen vs. eficiencia

El análisis muestra una diferencia clara: Shein publica una cantidad muy alta de contenido (200 posts frente a 28 de Temu), pero su eficiencia es tan baja —y además imposible de calcular por falta de datos— que cada publicación genera poca o ninguna interacción proporcional. Esto podría indicar que su audiencia está saturada o que el volumen excesivo diluye la atención que reciben sus posts.
No obstante, también puede tratarse de una estrategia totalmente intencional: Shein podría priorizar presencia constante sobre engagement por pieza, apostando por dominar el feed en lugar de generar interacciones profundas.
Temu, en cambio, parece apostar por un modelo de “menos es más”: publica poco, pero con un engagement consistente y relativamente alto, lo que indica una estrategia más cuidada y eficiente.
Aunque no podemos afirmar de forma absoluta que Shein “desgasta” a su audiencia, los datos sí apuntan a que su enfoque de cantidad no se traduce en eficiencia por publicación, mientras que Temu logra mejores resultados con una frecuencia mucho menor.

#### Greenwashing detector
Si alguna marca publica un volumen relevante de contenido “eco” pero con engagement por debajo del promedio, eso sugiere que la audiencia no lo percibe como auténtico sino como discurso superficial. El riesgo reputacional es alto: perder credibilidad, generar cinismo hacia la marca y activamente estimular backlash en redes. En sectores fast-fashion, donde la sostenibilidad ya es un tema sensible, un mal performance en contenido “eco” suele indicar riesgo de greenwashing percibido.

Esta combinación maximiza retorno por dólar invertido y fortalece ventaja competitiva.
